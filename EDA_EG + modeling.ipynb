Data cleaning and feature engineering

Making a copy of dataframe

```python
df_p = df.copy()
df_p.head()
```

Removing "Travellers Cheque" adding "Other" instead

```python
df_p['payment_mode'] = df_p['payment_mode'].replace('Travellers Cheque', 'Other', regex=True)
df_p['payment_mode'].value_counts()
```

Changing tapes of total_male(female) + adding new column

```python
df_p['total_female'] = df_p['total_female'].fillna(df_p['total_female'].value_counts().argmax())
df_p['total_male'] = df_p['total_male'].fillna(df_p['total_male'].value_counts().argmax())
index_values = df_p[(df_p["total_male"] == 'NaN') | (df_p["total_female"] == 'NaN')].index
df_p.drop(index_values, inplace=True)
df_p = df_p.astype({"total_female": int, "total_male": int})
df_p['group_size'] = df_p['total_female'] + df_p['total_male']
df_p['group_size']
```

remove "tour_", we agreed not to use this column for train data set

```python
#df_p['ID'] = df_p['ID']
df_p['ID'] = df_p['ID'].replace('[a-zA-Z_]', '', regex=True)
```

```python
display(df_p['ID'].duplicated())
```

```python
dummies = pd.get_dummies(df_p["travel_with"], prefix = "dummy")
dummies = dummies.astype(int)
# Concatenate the dummies DataFrame with the original DataFrame
df_p = pd.concat([df_p, dummies], axis=1)

# Display the updated DataFrame
dummies.head()
```

We have duplicates and typos on the list of countries let is removed it before transforming to numerical values

```python
df_p['country'] = df_p['country'].replace('DRC', 'CONGO', regex=True)
df_p['country'] = df_p['country'].replace('MALT', 'MALTA', regex=True)
df_p['country'] = df_p['country'].replace('BURGARIA', 'BULGARIA', regex=True)
df_p['country'] = df_p['country'].replace('UAE', 'UNITED ARAB EMIRATES', regex=True)
df_p['country'] = df_p['country'].replace('UKRAIN', 'UKRAINE', regex=True)

```

```python
dummies = pd.get_dummies(df_p["country"], prefix = "dummy")
dummies = dummies.astype(int)
# Concatenate the dummies DataFrame with the original DataFrame
df_p = pd.concat([df_p, dummies], axis=1)

# Display the updated DataFrame
dummies.head()
```

```python

# Evaluation of the baseline regression model on the training set
y_train_pred = lin_reg.predict(X_train)

train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)  
train_r2 = r2_score(y_train, y_train_pred)

print("Training Mean Absolute Error: {:.2f}".format(train_mae))
print("Training Mean Squared Error: {:.2f}".format(train_mse))
print("Training Root Mean Squared Error: {:.2f}".format(train_rmse))
print("Training R-squared: {:.2f}".format(train_r2))
```
